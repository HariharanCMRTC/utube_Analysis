```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("USvideos.csv")
```

```python
genre = pd.read_json("US_category_id (1).json")
```

```python
df.drop(["tags","thumbnail_link","description"],axis = 1, inplace = True)
```

```python
df["trending_date"] = pd.to_datetime(df["trending_date"], format = "%y.%d.%m")
```

```python
df["publish_time"] = pd.to_datetime(df["publish_time"])
```

```python
genre
```

```python
title = {}
for i in genre['items'].values:
  j = i['snippet']['title']
  title[i['id']]= [j]
```

```python
genre = pd.DataFrame({'category_id':title.keys(),'Genre':title.values()})
```

```python
genre['Genre']=[i[0] for i in genre['Genre'].values]
```

```python
genre['category_id']=pd.to_numeric(genre['category_id'])
```

```python
genre
```

```python
data = pd.merge(df,genre,on='category_id',how='inner')
```

```python
data
```

```python
data.to_csv("cleaneddata.csv")
```
